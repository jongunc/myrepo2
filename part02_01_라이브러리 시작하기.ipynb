```python
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,14,100) #linspace(시작값,끝값, 개수)
x
```

```python
y = np.sin(x) # 100개의 x 값에 대해서 100개의 sin값을 구해서 y로 넣어준다
plt.plot(x,y)
```

```python
y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4

# 그래프 그리기
plt.plot(x, y1, x, y2, x, y3)
```

```python
y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4

# 그래프 그리기
plt.plot(x, y1,  label='sin(x)')
plt.plot(x, y2,  label='sin(x)*2')
plt.plot(x, y3,  label='sin(x)*4')
plt.legend()
plt.title("sin plot")
```

```python
y1 = np.cos(x)
y2 = np.cos(x) * 2
y3 = np.cos(x) * 4

# 그래프 그리기
plt.plot(x, y1,'o',  label='sin(x)')
plt.plot(x, y2,'^',  label='sin(x)*2')
plt.plot(x, y3,  label='sin(x)*4')
plt.legend()
plt.title("cos plot")
```
